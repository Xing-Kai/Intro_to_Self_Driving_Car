```python
# open matrix.py (File > Open...) and start
# implementing matrix methods! 

# when your code passes all the tests you can submit by 
# pressing the SUBMIT button in the lower right corner 
# of this page.
import matrix as m


I2 = m.Matrix([
        [1, 0],
        [0, 1]
        ])
I2_neg = m.Matrix([
        [-1, 0],
        [0, -1]
        ])

zero = m.Matrix([
        [0,0],
        [0,0]
        ])

m1 = m.Matrix([
        [1,2,3],
        [4,5,6]
        ])

m2 = m.Matrix([
        [7,-2],
        [-3,-5],
        [4,1]
        ])

m1_x_m2 = m.Matrix([
        [ 13,  -9],
        [ 37, -27]])

m2_x_m1 = m.Matrix([
        [ -1,   4,   9],
        [-23, -31, -39],
        [  8,  13,  18]])

m1_m2_inv = m.Matrix([
        [1.5, -0.5],
        [2.0555556, -0.722222222]
        ])

top_ones = m.Matrix([
        [1,1],
        [0,0],
        ])

left_ones = m.Matrix([
        [1,0],
        [1,0]
        ])

def equal(m1, m2):
    if len(m1.g) != len(m2.g): return False
    if len(m1.g[0]) != len(m2.g[0]): return False
    for r1, r2 in zip(m1.g, m2.g):
        for v1, v2 in zip(r1, r2):
            if abs(v1 - v2) > 0.0001:
                return False
    return True
```
